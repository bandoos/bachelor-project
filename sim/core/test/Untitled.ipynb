```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
stakes = np.array([1,2,3,4])
```

```python
def txfun(m,a,b):
    return lambda x: m*x - b*(np.exp(a*x)-1)

```

```python
tx114 = np.vectorize(txfun(1,0.8,0.6))
tx111 = np.vectorize(txfun(1,.2,0.6))
xs = np.linspace(0,1)
ys = tx111(xs)
plt.plot(xs,np.log(xs+1),color="red")
plt.plot(xs,xs,color="blue")
plt.plot(xs,tx114(xs),color="orange")
plt.plot(xs,tx111(xs),color="green")
```

```python
res = tx114(stakes/np.sum(stakes))
res/np.sum(res)
```

```python
# Exp decays

def make_exp(decay):
    return np.vectorize(lambda x: np.exp(-decay*x))

exp_1 = make_exp(0.001)
exp_2 = make_exp(0.002)

xs = np.linspace(0,1000,num=1000)
plt.plot(xs,exp_1(xs),color="red")
plt.plot(xs,exp_2(xs))
```
